```python
import tensorflow as tf
import visualkeras
from tensorflow import keras
from keras import datasets, layers, models
from keras import backend as K
K.clear_session()
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)

model = keras.Sequential([
  layers.Rescaling(1./244, input_shape=(224, 224, 3)),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(96, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(256, activation='relu'),
  layers.Dense(8)
])

visualkeras.layered_view(model, legend=True, to_file='model.png').show()
tf.keras.utils.plot_model(model, show_shapes=False, to_file='model2.png', show_layer_names=False, show_layer_activations=True)
```
